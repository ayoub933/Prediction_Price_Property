```python
# Répartition des annonces par ville/adresse
requête = """
SELECT address, COUNT(*) AS nb_annonces
FROM properties
GROUP BY address
ORDER BY nb_annonces DESC;
    """
cursor.execute(requête)
resultats = cursor.fetchall()

for ligne in resultats:
    print(ligne)
```

```python
# Prix minimum, maximum et moyen
requête = """
SELECT 
    MIN(price) AS prix_min,
    MAX(price) AS prix_max,
    AVG(price) AS prix_moyen
FROM properties;
    """
cursor.execute(requête)
resultats = cursor.fetchall()

for ligne in resultats:
    print(ligne)
```
